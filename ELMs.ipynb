<center><h4>Abstract</h4></center>
<p>This thesis presents methods to detect edge localised modes (ELMs) in tokamak plasmas and uses these for the analysis of ELM characteristics. ELMs are plasma eruptions at the outer edge of tokamak plasmas that exert large heat loads on the plasma facing components. Two detection mechanisms were developed in the gas puff imaging (GPI) and outer divertor current (Ipolsola) diagnostics and their accuracy tested, showing lowest false positive rates of $\sim 3\%$ in the GPI and $\sim 4\%$ in the Ipolsola. Using these mechanisms the temporal ELM shape was averaged over 144 (GPI) and 163 ELMs (Ipolsola).
The coefficient of variation of the ELM period of $36\%$ (GPI) and $29\%$ (Ipolsola) confirms the quasi-periodicity of the ELMs described in literature. 
Finally, an apparent time offset between the two diagnostics signals was investigated, showing not to be of physical nature but rather a mismatch in the involved clocks.
</p>

<h3>1. Introduction</h3>
<p>  
When deuterium and tritium are fusing together to form an alpha particle, a neutron and excess energy of ${17.6}{MeV}$ are released in the process [1]. In order for fusion to take place the two nuclei must overcome each other's Coulomb barrier to enter the range of nuclear separation where the strong nuclear force is dominant. One method to achieve this is to heat up the deuterium and tritium until their thermal energies are large enough to overcome their Coulomb barriers. This process is called thermonuclear fusion and requires mean temperatures of $k_{B} T = 10keV$ [2]. At these temperatures the deuterium and tritium atoms are fully ionised, defining a new class of gases called plasma [3]. Due to the high temperatures of the plasma, it is not possible to confine the plasma within material walls. A tokamak reactor, that has the shape of a torus, confines the plasma using a combination of strong poloidal and toroidal magnetic fields without the plasma coming into contact with the material walls [1]. <br>
    
Even though the plasma's net electric charge is neutral, it consists of an equal amount of positively charged ions and free electrons. Hence, the plasma dynamics are mostly due to the electromagnetic interaction between the particles and with the external magnetic field [3]. 
The strong toroidal magnetic field allows for high velocities of the plasma particles along the magnetic field lines, but restricts the radial motion to a gyrating motion and drift velocities of much smaller magnitudes and therefore confines the plasma in a toroidal shape [3].
The drift velocities are of opposite direction for ions and electrons, leading to unwanted charge separation. Hence, a toroidal plasma current is induced by transformer coils located in the centre of the torus. This plasma current produces a poloidal magnetic field which combined with the externally applied toroidal magnetic field results in helical magnetic field lines that hinder charge separation and improve the confinement. 
The external toroidal field is typically in the order of a few teslas, whereas the poloidal field is usually an order of magnitude smaller and associated with a plasma current in the MA range [1,2]. The helical field lines create closed magnetic flux surfaces that contain the plasma. 
The last closed flux surface (LCFS), also described as the separatrix, describes the boundary between the core plasma and the so called scrape-off layer (SOL) [4]. The SOL is the region at the outer edge of the tokamak where the magnetic field lines are open. Thus, they are not forming closed surfaces but instead reach plasma facing components as seen in figure 2. The SOL is characterised by its low temperature and low plasma density compared to the core plasma [4]. <br>
    
There are two fundamentally different approaches to understand the plasma dynamics in an electromagnetic field. One deriving the overall plasma behaviour from the single-particle motion in an externally applied electric and magnetic field. This approach is valid for plasmas of low densities where the interaction between the plasma particles can be mostly neglected [3]. When studying slow and large scale phenomena in high density and high temperature plasmas that are confined by strong magnetic fields as present in a tokamak, it becomes more useful to use statistical methods by introducing a particle distribution function $f(\vec{r},\vec{v},t)$ and to treat the plasma as a single fluid. This approach is called magnetohydrodynamics (MHD) [3].
Edge localised modes (ELMs) are an MHD instability that become apparent when the tokamak has transitioned to the high confinement regime (H-mode) [5]. 
Whereas the low confinement regime (L-mode) is characterised by a comparatively low temperature gradient and significant particle transport through the LCFS, the H-mode is achieved by an increase in plasma heating power and leads to the development of an edge transport barrier (ETB) just inside the LCFS [4,6]. The ETB is characterised by high gradients in both plasma density and temperature [7]. <br>
    
The rising temperature gradient results in an ELM instability where plasma eruptions are transported into the SOL of the tokamak and then follow the helical magnetic field lines to the divertor and other plasma facing components (PFC) at the outer edge of the tokamak. The ELM instability leads to a decrease in the density and temperature gradients of the ETB which then build up until the next ELM instability is triggered [7]. <br>
    
An intrinsic feature of H-mode is the confinement of fuel-ions and impurities in the core plasma [7]. This is caused by the ETB as it prevents particle transport from the core plasma to the SOL. The impurities lead to radiative energy loss due to bremsstrahlung arising from the interaction of plasma particles with high Z (atomic number) impurities. ELMs are helpful in this regard as they help to transport impurities out of the core plasma through the ETB into the SOL [7]. <br>
        
However due to the short ELM duration of $\sim {200}{\mu s}$ [8] the high heat flux can result in damage of PFCs and diagnostics in future large tokamaks such as ITER [7]. The energy of each ELM and the associated heat load can be reduced by increasing the ELM frequency (ELM mitigation) or by finding mechanisms to prevent ELM instabilities (ELM suppression) [9]. Both ELM mitigation and ELM suppression require a better understanding of ELMs for which detection methods and knowledge of their characteristics are needed. A key challenge in detecting ELMs is their different shapes and quasi-periodicity. <br>
    
Section 2.1 describes the diagnostics in which the detection and analysis of the ELMs takes place and introduces additional plasma parameters, while section 2.2 described the working principles of the two mechanisms that were developed to detect the ELMs.
The first part of the results 3.1 analyses the detection accuracy of the mechanisms. The second part 3.2 analyses the temporal shape of an ELM, their periodicity and a time offset that was found between the GPI and Ipolsola data.
</p>

<p>An important characteristic of a magnetically confined fusion configuration is its safety factor which is generally defined as the inverse of the rotational transform $\iota$
\begin{equation}
    q \equiv \frac{2 \pi}{\iota}         \tag{1}
\end{equation}
where the rotational transform is given as the change in the poloidal angle $\vartheta$ after a full toroidal transit of the helical formed magnetic field lanes:
\begin{equation}
    \iota(r) \equiv \Delta \vartheta(r) = \vartheta(\varphi + 2 \pi) - \vartheta(\varphi)  \tag{2}
\end{equation}
with the toroidal angle $\varphi$ and the tokamak major radius $R$ [1, pp.286-287]. When approximating the torus as a cylinder this becomes:
\begin{equation}
    \iota(r) \approx 2\pi \frac{R B_{p}(r)}{r B_{t}(r)}     \tag{3}
\end{equation}
with the poloidal magnetic field $B_{p}$ and the toroidal magnetic field $B_{t}$. Hence, the safety factor is given by [1]
\begin{equation}\label{q-value}
    q(r) = \frac{2 \pi}{\Delta \vartheta(r)} \approx \frac{r B_{t}(r)}{R B_{p}(r)}    \tag{4}
\end{equation}
The first step of equation \ref{q-value} best illustrates that the q-value can be understood as a measure of how many toroidal transits of the magnetic field lanes are necessary in order to achieve a full poloidal transit, i.e. $\vartheta = 2 \pi$. 
The name safety factor originates from the fact that plasmas with larger q-values are more stable against MHD instabilities [1]. An often used q-value is $q_{95}$ which represents the q-value at $95\%$ of the magnetic flux [13,14]. 
</p>

<h5>2.1.3 ELM Dynamics</h5>
<p>The speed of sound in a fluid is generally given by the Newton-Laplace equation [15]:
\begin{equation}\label{eq: Cs}
    C_s = \sqrt{B/\rho}     \tag{5}
\end{equation}
with the bulk modulus B and fluid density $\rho$. 
The bulk modulus B is generally given by [16, p.535]:
\begin{equation}
    B = -V \frac{dp}{dV}      \tag{6}
\end{equation}
On the length scale that the sound waves propagate it is valid to treat the plasma as an ideal gas whose inertia is mainly given by the plasma ions ($m_i \gg m_e$) [3, p.190]. In this approximation the fluid density becomes:
\begin{equation}
    \rho = n_{i} m_{i}        \tag{7}
\end{equation}
By using the ideal gas law $pV = Nk_{B}T$ and assuming a constant plasma temperature, i.e. $pV = const$, the bulk modulus is then given by:
\begin{equation}
    B_T = -V \frac{d}{dV}\left(\frac{ N_{i}k_{B}T}{V}\right) = n_{i}k_{B}T   \tag{8}
\end{equation}
where $k_B$ is the Boltzmann constant. 
Hence, by using equation \ref{eq: Cs} the ion speed of sound is given by:
\begin{equation}
    C_s = \sqrt{\frac{n_{i}k_{B}T}{n_{i} m_i}} = \sqrt{\frac{2k_{B}T}{5m_p}}    \tag{9}
\end{equation}
where the ion mass $m_i = 5m_p/2$ is approximated by the mean of the deuterium ion mass of $\sim 2m_p$ and tritium ion mass of $\sim 3m_p$, where $m_p$ is the proton mass.   
</p>

<h5>2.2.2 ELM frequency</h5>
<p>A mechanism was developed to assign one frequency $f_\mathrm{ELM}$ to a given time history of the GPI or Ipolsola data. This was done by setting the frequency that corresponds to the peak in the power spectrum of the signal as the ELM frequency. The power spectrum was calculated by taking the Fourier transform of the autocorrelation function (ACF) of the signal [20, p.94]. The idea was that the ELM frequency is the most dominant frequency in the ACF. <br>
The discrete autocorrelation function of the discrete real function s[t] is given by
\begin{equation}
    \mathrm{ACF}(s[t])(\tau) =  \sum_{t=0}^{n-1} s[t+\tau] * s[t]    \tag{10}
\end{equation}
$\tau$ was chosen to run between $[-\frac{n-1}{2},\frac{n-1}{2}]$ for odd n and between $[-\frac{n}{2},\frac{n}{2}-1]$ for even n such that the resulting function has a length of n. Note that any terms where one of the indices runs outside the range of s[t] is set to zero. <br>
The discrete Fourier transform (DFT) of the ACF is given by:
\begin{equation}
    s[k] = \sum_{\tau=0}^{n-1} \mathrm{ACF}[\tau] \exp \left\{-i 2\pi\frac{k}  {n}\tau\right\}
\tag{11}
\end{equation}
where k=0,..,n-1.<br>
The ACF was normalized by the number of overlaps included in order to reduce the low frequency component</p>

```python
def detectPuffs(arraylength=500,threshold=0.05,FlagDelta=int(0.5/5e-6)):
    t0idx, t1idx = 0, time.shape[0]
    x = arraylength
    f_tilde = []
    for i in range(t0idx,t1idx-2*x):
        y = np.sum(frames[i+x:i+2*x,50,30],axis=0) - np.sum(frames[i:i+x,50,30],axis=0) #puffs detected at R=50px
        if y>0:
            f_tilde.append(y)
        else:
            f_tilde.append(0)
    maxvalue = np.max(f_tilde)
    f_tilde = [i/maxvalue for i in f_tilde] #normalize f_tilde
    thres = threshold
    f2 = []
    for i in f_tilde:
        if i >= thres:
            f2.append(i)
        else:
            f2.append(0)

    #mark start and final idx of each peak in gradDetectionResult
    truncStart = []
    truncFinal = []
    for i in range(1,len(f2)):
        if (f2[i] != 0 and f2[i-1] == 0):
            truncStart.append(i)
        if (f2[i] == 0 and f2[i-1] != 0):
            truncFinal.append(i)

    if len(truncStart) != len(truncFinal):
        print("error: truncation arrays not same size, check if time series starts or/and ends on peak")
        print("the lengths are: len(truncStart) = "+str(len(truncStart))+"; len(truncFinal) = "+str(len(truncFinal)))
    else:
        #correcting for too many flags in one area which are triggered by "sharp" threshold
        truncStart2 = []
        truncFinal2 = []
        Startidx = []

        for i in range(1,len(truncStart)):
            if (i == 1):
                truncStart2.append(truncStart[0]) #the first start item is valid by definition of this filtering
            if (truncStart[i]-FlagDelta > truncStart[i-1]):
                truncStart2.append(truncStart[i]) #item only appended if it is a "true" start item
                Startidx.append(i) #collection of the used idx of truncStart
        for k in Startidx:
            #the final item before each start item is the final item to the predecessor start item
            truncFinal2.append(truncFinal[k-1])

        truncFinal2.append(truncFinal[-1]) #the last Final item is valid by definition of this filtering
        # *** correction end ***

        if len(truncStart2) != len(truncFinal2):
            print("error: corrected truncation arrays are not the same size")
            print("the lengths are: len(truncStart2) = "+str(len(truncStart2))+"; len(truncFinal2) = "+str(len(truncFinal2)))
        else:
            timeStartResult = [time[i] for i in truncStart2]
            timeFinalResult = [time[i] for i in truncFinal2]
            print(len(truncStart2)," puffs detected with the following start and final times")
            print("Start timestamps: ",timeStartResult)
            print("Final timestamps: ",timeFinalResult)
            return np.array([truncStart2,truncFinal2]) #start and final indices
```

```python
def detectIpolsolaPuffs():
    Ipuffs = []
    starts = []
    finals = []
    for i in range(puffs.shape[1]):
        start_idx = np.argwhere(Ipolsola["time"] <= time[puffs[0,i]])[-1][0]
        final_idx = np.argwhere(Ipolsola["time"] >= time[puffs[1,i]])[0][0]
        starts.append(start_idx)
        finals.append(final_idx)
    Ipuffs.append(starts)
    Ipuffs.append(finals)
    Ipuffs = np.array(Ipuffs)
    return Ipuffs

IpolsolaPuffs = detectIpolsolaPuffs()
```

```python
def gradDetection(puff_no,diagnostic,R=55,arraylength=100,deltaIdx=[0,0],FlaggingThreshold=0.25, FlagDelta=300):
    
    validDiagn = {"GPI","Ipolsola"}
    if diagnostic not in validDiagn:
        raise ValueError("results: diagnostic must be one of %r." % validDiagn)
        
    validR = {55,22}
    if R not in validR:
        raise ValueError("results: R must be one of %r." % validR)
    
    if diagnostic == "GPI":
        t0idx, t1idx = puffs[0,puff_no-1]+deltaIdx[0], puffs[1,puff_no-1]+deltaIdx[1]
        if R == 55:
            absGrad = np.absolute(np.gradient(framesAVG_R55[0:t1idx]))
        if R == 22:
            absGrad = np.absolute(np.gradient(framesAVG_R22[0:t1idx]))
    elif diagnostic == "Ipolsola":
        t0idx, t1idx = IpolsolaPuffs[0,puff_no-1]+deltaIdx[0], IpolsolaPuffs[1,puff_no-1]+deltaIdx[1]
        absGrad = np.absolute(Ipolsola["data"][0:t1idx])  #the raw Ipolsola data is already similar to the grad of GPI
    
    # step 1 start
    f = [] 
    x = arraylength
    for i in range(t0idx,t1idx-2*x):
        f_i = np.sum(absGrad[i+x:i+2*x]) - np.sum(absGrad[i:i+x]) #foward difference of moving average i-th element
        if f_i>0:
            f.append(f_i) 
        else:
            f.append(f_i)
    
    maxvalue = np.max(f) 
    f = [i/maxvalue for i in f] 
    # step 1 end
    
    flags = flagELM(f=f,FlaggingThreshold=FlaggingThreshold,FlagDelta=FlagDelta)   
    
    return [f,flags[1],np.add(t0idx+arraylength,flags[2])]

def flagELM(f,FlaggingThreshold=0.5,FlagDelta=300):
    t0idx = 0
    thres = FlaggingThreshold
    f2 = []
    x = 0 #offset: the 0th conv element is the 0th signal element overlapping woth 0th ELM element
    
    for i in f:
        if i >= thres:
            f2.append(i)
        else:
            f2.append(0)
    #mark start and final idx of each peak in gradDetectionResult
    truncStart = []
    truncFinal = []
    for i in range(1,len(f2)):
        if (f2[i] != 0 and f2[i-1] == 0):
            truncStart.append(i)
        if (f2[i] == 0 and f2[i-1] != 0):
            truncFinal.append(i)

    if len(truncStart) != len(truncFinal):
        print("error: truncation arrays not same size, check if time series starts or/and ends on peak")
        print("the lengths are: len(truncStart) = "+str(len(truncStart))+"; len(truncFinal) = "+str(len(truncFinal)))
    else:
        #correcting for too many flags in one area which are triggered by "sharp" threshold
        truncStart2 = []
        truncFinal2 = []
        Startidx = []

        for i in range(0,len(truncStart)):
            if (i == 0): 
                truncStart2.append(truncStart[i]) #the first start item is valid by definition of this filtering
            else: 
                if (truncStart[i]-FlagDelta > truncStart[i-1]):
                    truncStart2.append(truncStart[i]) #item only appended if it is a "true" start item
                    Startidx.append(i) #collection of the used idx of truncStart
        for k in Startidx:
            #the final item before each start item is the final item to the predecessor start item
            truncFinal2.append(truncFinal[k-1])

        truncFinal2.append(truncFinal[-1]) #the last Final item is valid by definition of this filtering
        # *** correction end ***

        if len(truncStart2) != len(truncFinal2):
            print("error: corrected truncation arrays are not the same size")
            print("the lengths are: len(truncStart2) = "+str(len(truncStart2))+"; len(truncFinal2) = "+str(len(truncFinal2)))
        else:
            flagCoords = []
            for j in range(len(truncStart2)):
                flagCoords.append(np.argmax(np.array(f2[truncStart2[j]:truncFinal2[j]]))+x+truncStart2[j]+t0idx)
                #+x to get to middle pixel idx,+truncStart[j] to correct for truncation shift
                #+t0idx to correct for gaspuffshift/overall interval shift
            return [f,f2,np.array(flagCoords)]
```

In the first step a moving average (MA) with a boxcar length of $\Delta t = {500}{\mu s}$ is created 
\begin{equation}
    \mathrm{MA}[t] = \sum_{n=0}^{\Delta t} s[t-n]     \tag{12}
\end{equation}
with $s[t]$ being the time derivative of the GPI signal or the original Ipolsola signal.
This is then used to compute the forward difference $f[t]$ [21, p.182]:
\begin{equation}\label{eq:f}
    f[t] = \mathrm{MA}[t+\Delta t] - \mathrm{MA}[t]      \tag{13}
\end{equation}
Note that MA is computed backwards such that $f$ is centred on t.
The idea here is that a local maximum of $f$ means that this point corresponds to the start of an ELM. 
The normalised output of this step is shown in figure 7.

```python
plt.figure(figsize=(10,6))
plt.plot(np.arange(0,len(f)*5e-6,5e-6),f)

plt.xlabel('time [s]',fontsize=14)
plt.ylabel(r'$f\: [t]$',fontsize=14)
plt.title(r'$f\: [t]$ based on 1st time derivative of GPI of shot #36341',fontsize=14)
plt.xlim(0,0.075)
#plt.savefig("plots/f_function.pdf")
```

<b>Figure 7:</b> The normalised forward difference ${f}$ defined in equation 13 plotted for a segment of the GPI data of \#36341 and normalised by the maximum of $f$.

An additional parameter of the algorithm is introduced that will be referred to as the _flagging threshold_. This parameter is necessary to decide which of the peaks visible in figure 7 will be accepted as an ELM by requiring a minimum height for a peak:
\begin{equation}\label{eq:f2}
    f_2 = 
\begin{cases}
  f[t] & \text{for }f[t] > \mathrm{threshold} \\    
  0    & \text{for }f[t]\le \mathrm{threshold}    
\end{cases}
\tag{14}
\end{equation}
In the next step, the algorithm would search for local maxima of $f_2$ by determining the maxima within each time interval during which all data points are consistently different from zero (referred to as _search intervals_):
\begin{equation}
    \max(f_2[t]) = f_2[t_{ELM}]     \tag{15}
\end{equation}
where t runs only within the i-th search interval $t \in [t_i^{\mathrm{start}},t_i^{\mathrm{final}}]$. <br>
However, by applying the <i>flagging threshold</i> to $f$ there is occasionally more than one search interval within one larger peak that seems to be caused by the same ELM. This phenomenon is displayed in figure 8 and stems from the sharp character of the threshold. Small fluctuations of ${f}$ around the threshold will cause multiple search intervals.

To correct for this, a second parameter is introduced that will be referred to as the _flagging delta_, which allows to merge some search intervals together into larger _valid intervals_. <br>
A search interval's start time is only a start time for a valid interval if the start time of the last valid interval is further away than _flagging delta_ (by definition the start time of the first search interval is a valid start time). 
Identifying a new valid interval's start time as the start time of the i-th search interval allows to determine the final time of the last valid interval as the final time of the last search interval $i-1$. <br>
In a final step the maximum of each valid interval is determined and the associated time saved as the start of one ELM.

<h5>2.2.4  Mechanism 2</h5>
<p> This mechanism makes use of the cross-correlation of the GPI signal and the time series of a single ELM. Note that compared to the convolution of $s$ and $g$, the cross-correlation does not flip $g$ before computing the overlap with $s$. Let $s[t]$ be the GPI signal and $g[t]$ the shorter time series of a single ELM region of length $\Delta t = {4.5}{ms}$. $g$ was chosen manually as an average sized ELM. The length $\Delta t$ is chosen as such to contain the section of increased intensity around an ELM. The ELM region sample $g[t]$ has a labelled ELM start time at $t_{sample}^{ELM}$. The discrete cross-correlation is then defined as:
\begin{equation}\label{eq:conv}
    c[\tau] = g[t] * s[t] = \sum_{t=0}^{\Delta t} g[\tau+t] \cdot s[t]
\tag{16}
\end{equation}
The idea is that $c[\tau]$ will have higher values for shift times $\tau$ where $s$ has a similar shape to $g$. $c$ generally shows no clear change for different $g$ as long as $g$ is chosen as the same type of ELM [5, p.107]. The output of this step can be seen in figure 10.</p>

The local maxima will then respond to the best overlap between both signals. The location of the ELM in $s[t]$ can be calculated as:
\begin{equation}
    t_i^{ELM} = \tau_i^{ELM} + t_{sample}^{ELM}
\tag{17}
\end{equation}
where $\tau = 0$ corresponds to the first complete overlap of both $s$ and $g$. As $c[\tau]$ has a similar shape as the function $f$ of Mechanism 1 in figure 6 of section 2.2.3, the remainder of this algorithm works analogous to Mechanism 1 with the use of a _flagging threshold_ and a _flagging delta_.
Note however, that generally the optimal values for these parameters will be different for both mechanisms.

```python
def convDetection(puff_no,diagnostic,FlaggingThreshold=0.5,FlagDelta=300,deltaIdx=[0,0]):
    
    validDiagn = {"GPI","Ipolsola"}
    if diagnostic not in validDiagn:
        raise ValueError("results: diagnostic must be one of %r." % validDiagn)
        
    if diagnostic == "GPI":
        elm = np.load("data/GPIconvSample.npy")
        flag_idx = 269
        t0idx, t1idx = puffs[0,puff_no-1]+deltaIdx[0], puffs[1,puff_no-1]+deltaIdx[1]
        conv = np.correlate(framesAVG_R55[t0idx:t1idx],elm,mode="full")
    elif diagnostic == "Ipolsola":
        elm = np.load("data/IpolConvSample.npy")
        flag_idx = 200
        t0idx, t1idx = IpolsolaPuffs[0,puff_no-1]+deltaIdx[0], IpolsolaPuffs[1,puff_no-1]+deltaIdx[1]
        conv = np.correlate(Ipolsola["data"][t0idx:t1idx],elm,mode="full") 
        
    conv = np.divide(conv,np.max(conv)) #normalize
    conv_time = np.arange(0,len(conv)*5e-6,5e-6) #create shift time tau array

    flaggedELMs = flagELM(f=conv,FlaggingThreshold=FlaggingThreshold,FlagDelta=FlagDelta)
    # correcting tau = 0 at puff beginning
    flaggedELMs = np.add(flaggedELMs[2],t0idx)
    
    #offset caused by right edges (re) of elm region lining up giving max in this algorithm
    #therefore correct the shift from right edge to actual flag
    re_idx = len(elm)-1
    ELMregion = [[i-re_idx,i+1] for i in flaggedELMs]
    flaggedELMs = np.add(flaggedELMs,flag_idx-re_idx)
   
    return [flaggedELMs.astype(int),ELMregion]
```

<h5>2.2.5 Parameter optimisation</h5>
<p>The goal of this step is to find the values of the <i>flagging threshold</i> and <i>flagging delta</i> for each mechanism which result in the highest detection efficiency. In order to find the optimal values, ELM labels were created for two gas puffs of the GPI diagnostic. This corresponds to 56 labelled ELMs during $\sim {0.4}{s}$. </p> 

```python
GPI_labels_puff2, GPI_labels_puff3 = np.load("data/GPI_labels_36341.npy",allow_pickle=True)
Ipolsola_labels_puff2, Ipolsola_labels_puff3  = np.load("data/Ipolsola_labels_36341.npy",allow_pickle=True)
```

<p>In the next step a loss function $L$ is defined which assigns a real number to each parameter pair of <i>flagging threshold</i> (thres) and <i>flagging delta</i> (fdelta):
$$\label{eq:loss}
    L(\mathrm{thres},\mathrm{fdelta}) = fn(\mathrm{thres},\mathrm{fdelta}) +fp(\mathrm{thres},\mathrm{fdelta})  \tag{18}
$$
with the false negative rate defined as:
\begin{equation}\label{eq: fn}
    fn = \frac{FN}{FN+TP}   \tag{19}
\end{equation}
and a modified definition of the false positive rate:
$$ \label{eq: fp adapted}
    fp = \frac{FP}{FP+TP}  \tag{20}
$$
with FP,FN and TP being the number of false positives, false negatives and true positives detected ELMs (flags), respectively.
In a next step, $L$ was evaluated for a range of both parameters to determine the parameter values with minimal loss values.
A labelled ELM was counted as being correctly detected if the flag raised by the mechanism lies within an interval of $\pm {1}{ms}$ of the label. It showed that flags within this region seem to refer to the same ELM.</p>

```python
def Errorrates(result,groundtruth,AcceptancedeltaIdx=200,withPrint=True):
    gt = groundtruth 
    delta = AcceptancedeltaIdx
    isELM = []
    isNotELM = []
    detected = []
    for i in result:
        isELMtest = False
        k = 0 
        for j in gt: 
            if (abs(i-j)<delta):
                isELMtest = True
                detected.append(j)
                # fix issue of negative fn values by not letting gt item be detected more than once
                gt = np.delete(gt,k)
                break
            k += 1
        if (isELMtest == True):
            isELM.append(i)
        else:
            isNotELM.append(i)
            
    tp = len(isELM) #is ELM and flagged as such
    fp = len(isNotELM) #is not ELM but flagged as such
    fn = len(groundtruth)-len(detected) #is ELM but not flagged as such
    
    if withPrint == True:
        print("TP: ",tp,",FP: ",fp,",FN: ",fn)
    
    fp_ratio = fp/(fp+tp) #false positives / total number of positives (against usual definition)
    fn_ratio = fn/(fn+tp) #false negatives / total number of true ELMs
    
    return [fp_ratio,fn_ratio]

def loss(puff2flags,puff2labels,puff3flags,puff3labels):
    e2 = Errorrates(puff2flags,puff2labels,withPrint=False)
    e3 = Errorrates(puff3flags,puff3labels,withPrint=False)
    fp_avg = np.mean([e2[0],e3[0]])
    fn_avg = np.mean([e2[1],e3[1]])
    print(fp_avg,fn_avg)
    l = fp_avg+fn_avg
    return l

def GPIconvParaOptimizer():
    minLoss = 1
    l = 0
    tresData = []
    for tres in np.arange(0.1,0.7,0.01):
        l += 100/60
        for flagd in np.arange(0,300,10):
            convPuff2flags, convPuff3region = convDetection(puff_no=2,diagnostic="GPI",FlaggingThreshold=tres, FlagDelta=flagd)
            convPuff3flags, convPuff3region = convDetection(puff_no=3,diagnostic="GPI",FlaggingThreshold=tres, FlagDelta=flagd)
            current_loss = loss(convPuff2flags,GPI_labels_puff2,convPuff3flags,GPI_labels_puff3)
            tresData.append(current_loss)
            if current_loss < minLoss:
                minLoss = current_loss
                tresOpt = tres
                flagdOpt = flagd
        print(l,"% done")          
    return [minLoss,tresOpt,flagdOpt,tresData]

def GPIgradParaOptimizer():
    minLoss = 1
    l = 0
    tresData = []
    for tres in np.arange(0.1,0.7,0.01):
        l += 100/60
        for flagd in np.arange(0,300,10):
            gradPuff2flags = gradDetection(puff_no=2,diagnostic="GPI",R=55,FlaggingThreshold=tres, FlagDelta=flagd)
            gradPuff3flags = gradDetection(puff_no=3,diagnostic="GPI",R=55,FlaggingThreshold=tres, FlagDelta=flagd)
            current_loss = loss(gradPuff2flags,GPI_labels_puff2,gradPuff3flags,GPI_labels_puff3)
            tresData.append(current_loss)
            if current_loss < minLoss:
                minLoss = current_loss
                tresOpt = tres
                flagdOpt = flagd
        print(l,"% done")          
    return [minLoss,tresOpt,flagdOpt,tresData]

def IconvParaOptimizer():
    minLoss = 1
    l = 0
    tresData = []
    for tres in np.arange(0.1,0.7,0.01):
        l += 100/60
        for flagd in np.arange(0,300,10):
            convPuff2flags, convPuff2region = convDetection(puff_no=2,diagnostic="Ipolsola",FlaggingThreshold=tres,FlagDelta=flagd,deltaIdx=[IpolsolaDeltaIdxStart,0])
            convPuff3flags, convPuff3region = convDetection(puff_no=3,diagnostic="Ipolsola",FlaggingThreshold=tres,FlagDelta=flagd,deltaIdx=[IpolsolaDeltaIdxStart+500,0])
            current_loss = loss(convPuff2flags,Ipolsola_labels_puff2,convPuff3flags,Ipolsola_labels_puff3)
            tresData.append(current_loss)
            if current_loss < minLoss:
                minLoss = current_loss
                tresOpt = tres
                flagdOpt = flagd
        print(l,"% done")          
    return [minLoss,tresOpt,flagdOpt,tresData]

def IgradParaOptimizer():
    minLoss = 1
    l = 0
    tresData = []
    for tres in np.arange(0.1,0.7,0.01):
        l += 100/60
        for flagd in np.arange(0,300,10):
            gradPuff2flags = gradDetection(puff_no=2,diagnostic="Ipolsola",FlaggingThreshold=tres, FlagDelta=flagd,deltaIdx=[IpolsolaDeltaIdxStart,0])
            gradPuff3flags = gradDetection(puff_no=3,diagnostic="Ipolsola",FlaggingThreshold=tres, FlagDelta=flagd,deltaIdx=[IpolsolaDeltaIdxStart+500,0])
            current_loss = loss(gradPuff2flags,Ipolsola_labels_puff2,gradPuff3flags,Ipolsola_labels_puff3)
            tresData.append(current_loss)
            if current_loss < minLoss:
                minLoss = current_loss
                tresOpt = tres
                flagdOpt = flagd
        print(l,"% done")          
    return [minLoss,tresOpt,flagdOpt,tresData]
```

<h5>2.2.6 Combining mechanisms</h5>
<p>In a final step, mechanism 1 (M1) of section 2.2.3 and mechanism 2 (M2) of section 2.2.4 were combined for both the GPI and the Ipolsola diagnostic. 
This combined mechanism (CM) allows to produce a single output of detected ELMs each assigned with a degree of confidence indicating whether the detection is based on both mechanisms, M1 only or M2 only.
As in section 2.2.5 two flags of both algorithms were accepted as referring to the same ELM if they are less than ${1}{ms}$ apart from each other.</p>

```python
def detectCombi(puff_no,diagnostic,R=55,arraylength_grad=100,deltaIdx_grad=[0,0],deltaIdx_conv=[0,0],FlaggingThreshold_grad="optimal", FlagDelta_grad="optimal",FlaggingThreshold_conv="optimal",FlagDelta_conv="optimal",deltaR_conv=17):
    
    validDiagn = {"GPI","Ipolsola"}
    if diagnostic not in validDiagn:
        raise ValueError("results: diagnostic must be one of %r." % validDiagn)

    if diagnostic == "GPI":
        gradOptParaTres, gradOptParaFlagd, convOptParaTres, convOptParaFlagd = 0.27, 260, 0.51, 240

        if FlaggingThreshold_grad == "optimal":
            FlaggingThreshold_grad = gradOptParaTres
        if FlagDelta_grad == "optimal":
            FlagDelta_grad = gradOptParaFlagd
        if FlaggingThreshold_conv == "optimal":
            FlaggingThreshold_conv = convOptParaTres
        if FlagDelta_conv == "optimal":
            FlagDelta_conv = convOptParaFlagd

        gradResult = gradDetection(puff_no=puff_no,diagnostic="GPI",arraylength=arraylength_grad,deltaIdx=deltaIdx_grad,FlaggingThreshold=FlaggingThreshold_grad,FlagDelta=FlagDelta_grad)[2] 
        convResultFlags, convResultRegion = convDetection(puff_no=puff_no,diagnostic="GPI",FlaggingThreshold=FlaggingThreshold_conv,FlagDelta=FlagDelta_conv,deltaIdx=deltaIdx_conv)

    elif diagnostic == "Ipolsola":
        gradOptParaTres,gradOptParaFlagd, convOptParaTres, convOptParaFlagd = 0.18, 290, 0.13, 90 

        if FlaggingThreshold_grad == "optimal":
            FlaggingThreshold_grad = gradOptParaTres
        if FlagDelta_grad == "optimal":
            FlagDelta_grad = gradOptParaFlagd
        if FlaggingThreshold_conv == "optimal":
            FlaggingThreshold_conv = convOptParaTres
        if FlagDelta_conv == "optimal":
            FlagDelta_conv = convOptParaFlagd

        gradResult = gradDetection(puff_no=puff_no,diagnostic="Ipolsola",arraylength=arraylength_grad,deltaIdx=deltaIdx_grad,FlaggingThreshold=FlaggingThreshold_grad,FlagDelta=FlagDelta_grad)[2] 
        convResultFlags, convResultRegion = convDetection(puff_no=puff_no,diagnostic="Ipolsola",FlaggingThreshold=FlaggingThreshold_conv,FlagDelta=FlagDelta_conv,deltaIdx=deltaIdx_conv)

    combiResult = []
    for i in gradResult:
        both = False
        l = 0
        for j,k in zip(convResultFlags,convResultRegion):
            if abs(i-j) < 200:
                both = True
                combiResult.append([i,j,k,"both"])
                convResultFlags = np.delete(convResultFlags,l)
                convResultRegion = np.delete(convResultRegion,l,axis=0)
                break
            l += 1
        if both == False:
            combiResult.append([i,0,[0,0],"grad"])
            
    #append rest of convResultFlags that are not both
    for j,k in zip(convResultFlags,convResultRegion):
        combiResult.append([0,j,k,"conv"])
    
    combiResult = np.array(combiResult,dtype=object)    
    return combiResult
```

Another function was created in order to filter the flags by the detection mechanism behind it. Note that filtering for M1 and M2 also includes the flags that were accepted by CM.

```python
def FilterByFlag(FLagidxes,method="both"):
    if method != "both":
        if method == "grad":
            idx = 0
        if method == "conv":
            idx = 1

        FLagidxes_filtered = []
        for i in FLagidxes:
            if i[idx] != 0 :
                FLagidxes_filtered.append(i)
        FLagidxes_filtered = np.array(FLagidxes_filtered)
    else:
        idx = 0 #added
        FLagidxes_filtered = []
        for i in FLagidxes:
            if i[0] != 0 and i[1] != 0:
                FLagidxes_filtered.append(i)
        FLagidxes_filtered = np.array(FLagidxes_filtered)

    return FLagidxes_filtered[:,idx] #added
```

Table 2 also shows that <i>M1 generally performs better than M2 in the Ipolsola</i>. Notice that the false positive rate of CM is higher than the false positive rate of M2. 
This is due to the fact that the false positive rate is not only influenced by FP but also by TP. Significantly lower TP values of CM can increase the false positive rate.  
However, the results in table 2 and figure 11 and 12 suggest that this effect is rather small. <br>
It is not clear whether the differences between M1 and M2 would hold for other shots as the above results from one shot. However, it shows that <i>CM generally adopts the higher false negative rate and lower false positive rate of either M1 or M2</i>. 
This effect is more likely to extend to other shots: even if differences in false positive and false negative rates between M1 and M2 vary during other shots, both mechanisms still need to agree on a detected ELM. 
For the analysis in section 3.2 this effect is advantageous as it merely means that fewer ELMs will be included in the analysis than there were in the data. Whereas a higher false positive rate would falsely include data. <br>
<i>For other purposes, such as removing ELMs from the diagnostics data, it might be more useful to accept any ELMs that were detected by either M1 or M2 or by both</i>. <h4>3.2 ELM characteristics</h4>
<h5>3.2.1 Spatial and temporal distribution</h5>
<p>To create an average temporal ELM evolution CM was run on seven GPI puffs of five different shots. For the GPI (Ipolsola), 230 (210) ELMs were detected in total of which 144 (163) were detected by CM, 71 (34) by M1 only and 15 (13) by M2 only. <br>
The mean ELM was obtained by averaging the signal at each point in time across all detected ELMs each of length $\Delta t$. The ELMs are extracted from the full time series of the GPI or Ipolsola diagnostic through the knowledge of $t_i^{ELM}$ and $t_{sample}^{ELM}$ as described in section 2.2.4. Similarly, the power normalised mean was obtained by scaling the signal at each point in time by the reciprocal of the total power $P_i$ of the associated i-th ELM, where P is given by:
\begin{equation}
    P_i = \sum_{t=0}^{\Delta t} \left({ELM_{i}[t]}\right)^2
\tag{21}
\end{equation}

```python
def IpolsolaOffset(GPIresult,Iresult,timeIdxOffset=0,FlagFilter="grad",GPIFilter="yes",IpolFilter="yes"): #Filter ADDED,searchData=50 deleted
    #determine a start offset to know what order of magnitude to search in
    #also assume that offset is positive, i.e GPI flags before Ipolsola flags
    #also based on fact that Ipolsola sometimes flags before GPI start flagging
    if GPIFilter == "yes":
        GPIidxes = FilterByFlag(Test3GPI,method=FlagFilter)
    elif GPIFilter == "no":
        GPIidxes = GPIresult
        
    if IpolFilter == "yes":
        Iidxes = FilterByFlag(Iresult,method=FlagFilter)
    elif IpolFilter == "no":
        Iidxes = Iresult
        
        
    GPIidxes = np.add(GPIidxes,timeIdxOffset)
    offsetIdx = [] #based on Ipolsola flags
    k = 0
    for i in Iidxes:
        smallerGPIidx = []
        for j in GPIidxes:
            if j<i:
                smallerGPIidx.append(j)
        if k == 0:
            if len(smallerGPIidx) > 0:
                offset = abs(max(smallerGPIidx)-i)
                offsetIdx.append(offset)
            else:
                offsetIdx.append(0)
        else:
            if len(smallerGPIidx) != 0:
                if Iidxes[k-1] > max(smallerGPIidx):
                    offsetIdx.append(0)
                else:
                    offset = abs(max(smallerGPIidx)-i)
                    offsetIdx.append(offset)
            else:
                offsetIdx.append(0)
        k += 1
    offsetIdx = np.array(offsetIdx)
    return offsetIdx
```

```python
def Ipol_rho_mean(shot_no,puff_no):
    divrho = np.load("data/divrho_"+str(shot_no)+"_"+str(puff_no)+"s.npz")
    bgr2_mean_rho = []
    bgr2_std_rho = []
    for i in range(0,16): #baugruppe 2 in figure 2 consists of 16 tiles
        mean_rho = np.mean(divrho["rho2"][:,i]) #avg over time of the gas puff for this tile
        std_rho = np.std(divrho["rho2"][:,i])
        bgr2_mean_rho.append(mean_rho)
        bgr2_std_rho.append(std_rho)
    bgr1_mean_rho = []
    bgr1_std_rho = []
    for i in range(0,22): #baugruppe 1 in figure 2 consists of 22 tiles
        mean_rho = np.mean(divrho["rho1"][:,i]) #avg over time of the gas puff for this tile
        std_rho = np.std(divrho["rho1"][:,i])
        bgr1_mean_rho.append(mean_rho)
        bgr1_std_rho.append(std_rho)
    bgr12_mean_rho = np.append(bgr2_mean_rho,bgr1_mean_rho) #avg over all tiles
    bgr12_std_rho = np.append(bgr2_std_rho,bgr1_std_rho)
    bgr12_totalmean_rho = np.mean(bgr12_mean_rho)
    bgr12_totalstd_rho = np.mean(bgr12_std_rho)
    return [bgr12_totalmean_rho,bgr12_totalstd_rho]

def R_alignCalc(bgr12_totalmean_rho):
    R_aligned = 0
    rho_aligned = []
    rho_delta = 0
    for i in range(0,64): #avg along all 64 R-pixels
        rhoAVG = np.mean(rhoH[i,:]) #avg along z for each r
        rhoSTD = np.std(rhoH[i,:])
        if i == 0:
            rho_delta = abs(bgr12_totalmean_rho - rhoAVG)
        else:
            if abs(bgr12_totalmean_rho - rhoAVG) < rho_delta:
                rho_delta = abs(bgr12_totalmean_rho - rhoAVG)
                rho_aligned = [rhoAVG,rhoSTD]
                R_aligned = i
    return [R_aligned,rho_aligned,rho_delta]
```

<center><b>Figure 15:</b> The GPI-Ipolsola offset for the same magnetic location for all 7 shots.</center>

Figure 15 shows that there is a consistent offset between the GPI and Ipolsola of $\Delta t = (2.17 \pm 0.15)\,{ms}$.
The $q_{95}$ value for all shots used for figure 15 is $q_{95} \approx 4.8$. 
Studies on JET have shown that after ELMs have crossed the separatrix they tend to cause a sound wave disturbance that follows the magnetic field lines [22, p.1445].
Once triggered in the SOL at the midplane, the sound wave will reach the divertor region after completing approximately a quarter of a full poloidal transit.
Using equation 4, the number of toroidal transits that are necessary to achieve $\Delta \theta = \pi/2$ is given by:
\begin{equation}
    N = \frac{q_{95}}{4}
\tag{22}
\end{equation}
Given that the toroidal circumference $C = 2 \pi R$, the toroidal distance travelled by the ELM can be approximated as:
\begin{equation}
    \Delta x_t = N \cdot C = \frac{\pi R\, q_{95}}{2} \approx {11.31}{m}
\tag{23}
\end{equation}
with $R \approx {1.5}{m}$ as seen in figure 2. The poloidal distance travelled can be approximated by the distance in z-direction between the outer midplane and the divertor $\Delta z \approx {1}{m}$ as seen in figure 2.
The total distance can then be approximated by:
\begin{equation}
    \Delta x \approx \sqrt{(\Delta x_t)^2 + (\Delta z)^2} \approx {11.35}{m}
\tag{24}
\end{equation}
When approximating the parallel propagation speed of the disturbance with the plasma speed of sound derived in section  2.1.3 for a typical plasma temperature of $k_{B}T = {10}{keV}$ [2], this becomes
\begin{equation}
    C_s = \sqrt{\frac{2k_{B}T}{5m_p}} \approx {619}{km/s}
\tag{25}
\end{equation}
Hence, if the sound wave disturbance follows the helical magnetic field lines, the expected delay between the GPI and Ipolsola is given by
\begin{equation}\label{eq: Ipolsola_delay}
    \Delta t = \frac{\Delta x}{C_s} \approx {18}{\mu s} 
\tag{26}
\end{equation}
The works of Manfredi <i>et al.</i> describe a delay of a few hundred microseconds between the ELM crossing the separatrix and the plasma pulse hitting the divertor plate [23, p.2].
Comparing this value and the value obtained in equation 26 to the measured median offset of $(2.17 \pm 0.15)\,{ms}$ shown in figure 15 shows that the measured offset is at least one order of magnitude larger than expected. Thus, this offset does not seem to have a physical origin and is most likely caused by an offset in the involved clocks of the GPI and Ipolsola diagnostics.

<h5>3.2.3 ELM frequency & Periodicity</h5>
<p>
Figure 16 shows the time $\Delta t$ that passes between each successive pair of ELMs as well as the mean $\overline{T}_{ELM}$ and standard deviation $\sigma$ for each of the seven included gas puffs for the GPI and the Ipolsola diagnostic. 
The coefficient of variation (CV) is given by
\begin{equation}\label{eq:cv}
    c_v = \frac{\sigma}{\overline{T}_{ELM}}     \tag{27}
\end{equation}
The total mean ELM period is given by $\overline{T}_{GPI} = {9.3}{ms}$ in the GPI and $\overline{T}_{Ipol} = {8.5}{ms}$ in the Ipolsola. The total mean standard deviation are given by $\sigma_{GPI} = {3.4}{ms}$ and $\sigma_{Ipol} = {2.5}{ms}$. Using equation 4, this leads to a mean coefficient of variation of $36\%$ for the GPI and $29\%$ for the Ipolsola.
A perfectly periodic behaviour would be associated with a coefficient of variation of zero, whereas aperiodic behaviour, i.e a uniform distribution of $\Delta t$ would be associated with a coefficient of variation of $1/\sqrt{3} \approx 58\%$ in the continuous limit (see appendix B for derivation).
Therefore, the CV values determined for the GPI and Ipolsola confirm the quasi-periodic behaviour of ELMs stated in the works of Connor [24] and Keilhacker <i>et al.</i> [25]. <br>
In addition, Figure 16 shows the ELM period $T_{ACF}$ that results from the fourier transform of the ACF of the signal described in section 2.2.2. It shows that $T_{ACF}$ is close to the mean $T_{ELM}$ value for five of the seven puffs and lies withing the $1\sigma$-interval for both diagnostics. For the same two puffs there is a strong deviation between $T_{ACF}$ and $T_{ELM}$ in both diagnostics. When looking at the data, the signal shows higher inter-ELM fluctuations [26] during these puffs possibly caused by a less stable confinement configuration. This causes more dominant frequencies in the power spectrum which could explain this deviation.  
However, this confirms that even though the ELMs show quasi-periodic behaviour, the method described in section 2.2.2 provides a good approximation for the mean ELM frequency in a given gas puff. </p>

<h3>4. Conclusion</h3>

Two mechanisms M1 and M2 were developed to detect ELMs. M1 operates with the moving average of the signal, M2 with the cross-correlation between the signal and a single ELM. Both mechanisms were optimised and tested against a labelled shot in the GPI and Ipolsola diagnostics; M1 shows false positive (negative) rates of $11.69\%$ ($9.52\%$) in the GPI and $4.17\%$ ($2.86\%$) in the Ipolsola, whereas M2 shows false positive (negative) rates of $4\%$ ($17.14\%$)in the GPI and $3.23\%$ ($8.57\%$) in the Ipolsola. It shows that M1 generally detects more aggressive than M2.
Both mechanisms were combined into a single mechanism CM that requires both M1 and M2 to detect an ELM in order to output the detection. CM shows to generally adopt the lower false positive rate but the higher false negative rate of M1 and M2. <br>
The ELMs temporal shape was averaged over detected ELMs in seven shots, which resulted in 144 total detected ELMs in the GPI and 163 in the Ipolsola. The comparison of the mean ELM with the power normalised mean ELM shows that the detected ELMs are similarly shaped in the GPI while ELMs of different powers are detected in the Ipolsola. While the intensity in the GPI diagnostic builds up within microseconds during an ELM, the Ipolsola current rises to its maximum value in $\sim {0.5}{ms}$. Following an ELM the GPI shows increased intensity values for $\sim {3}{ms}$ while the Ipolsola shows an average time of increased current of $\sim {1}{ms}$. Future convolution based detection mechanisms could make use of the average ELM shapes and the uncertainty in amplitude at each point in time. <br>
An apparent offset between the ELMs detected in the GPI and the ones detected in the Ipolsola shows to have a median value of $(2.17 \pm 0.15)\,{ms}$. Comparisons with calculations based on the ion sound speed as well as delay values described in the works of Manfredi <i>et al.</i> [23] show that this offset is not of physical origin but caused by a mismatch in the involved clocks. <br>
A mechanism that assigns one ELM period to an entire time series of either diagnostic based on the Fourier transform of the autocorrelation function shows to be a good first approximation as it lies within the standard deviation of the mean ELM period for five out of seven GPI gas puffs.
The mean coefficient of variation (CV) of the time passed between two successive ELMs shows values of $36\%$ in the GPI and $29\%$ in the Ipolsola. As a normal distributed ELM period would lead to a CV of $\sim 58\%$, this confirms the quasi-periodic ELM behaviour stated in works of Connor [24] and Keilhacker <i>et al.</i> [25]. <br>
The detection mechanisms presented in this thesis provide an algorithmic basis for the development of detection mechanisms in other diagnostics such as the B-dot probe-based diagnostic [27]. In addition, they can be used for further studies of ELM characteristics such as precursor oscillations or the spatial mode structure.

<h4>B Uniform Distribution</h4>
<p>In the continuous limit the probability density function of a uniform distribution that has values different from zero between A and B is given by:
\begin{equation}
    f(x) = \frac{1}{B-A}
\end{equation}
with the mean given by
\begin{equation}
    \mu = \int_{-\infty}^{\infty} f(x)x \, dx  = \frac{B+A}{2}
\end{equation}
The variance is after a rather lengthy calculation [21] given by
\begin{equation}
    \sigma^2 = \langle x^2 \rangle - \mu^2 = \frac{(B-A)^2}{12}
\end{equation}
Hence the coefficient of variation is given by
\begin{equation}
    c_v = \frac{\sigma}{\mu} = \frac{(B-A)/\sqrt{12}}{(B+A)/2} = \frac{B-A}{\sqrt{3}(B+A)}
\end{equation}
In the case where $A = 0$ this simplifies to
\begin{equation}
    c_v = \frac{1}{\sqrt{3}}
\end{equation}
</p>
