<h3>1. Introduction</h3>
<p>  
When neutral gases are raised to a sufficiently high temperature most of the particles become ionised, defining a new class of gases called plasma [1].
The Coulomb forces in plasmas are particularly strong as the individual gas particles have a non zero net electric charge [1]. This non zero electric charge allow for the plasma to conduct electricity and to interact with magnetic field via the Lorentz force $q\mathbf{v} \times \mathbf{B}$ [1].

Tokamaks use strong toroidal magnetic fields in addition to poloidal fields created by the plasma current to confine the plasma in the shape of a torus [2]. The toroidal field is typically in the order of a few Tesla, whereas the poloidal field is usually an order of magnitude smaller and driven by a plasma current in the MA range [3]. <br> 

The confinement of the plasma in a tokamak is achieved by helical magnetic field lines [3] which lead to closed magnetic flux surfaces that contain and shield the core plasma. The last closed flux surface (LCFS), also described as the separatrix, describes the boundary between the core plasma and the so called scrape-off layer (SOL) [4]. The SOL is the region at the outer edge of the tokamak where the magnetic field lines are open; It is characterised by its low temperature and low plasma density compared to the core plasma [4].  <br>
Whereas the low confinement mode (L-mode) is characterised by a comparatively low temperature gradient and significant particle transport through the LCFS, the high confinement mode (H-mode) is achieved by an increase in plasma heating power and leads to the development of an edge transport barrier (ETB) just inside the LCFS [4,5]. The ETB is characterised by high gradients in both plasma density and temperature [6]. The H-mode leads to improved confinement of the plasma [4] but also gives rise to a plasma instability called edge-localised modes (ELMs).
The rising temperature gradient results in an ELM instability where plasma eruptions are transported into the SOL of the tokamak and then follow the helical magnetic field lines to the divertor and other plasma facing components (PFC) at the outer edge of the tokamak. The ELM instability leads to a decrease in the density and temperature gradients of the ETB which then build up until the next ELM instability is triggered [6]. 
An intrinsic feature of H-mode is the confinement of fuel-ions and impurities in the core plasma [6]. This is caused by the ETB as it prevents particle transport from the core plasma to the SOL. The impurities lead to radiative energy loss due to bremsstrahlung arising from the interaction of plasma particles with high Z (atomic number) impurities. ELMs are helpful in this regard as they help to transport impurities out of the core plasma through the ETB into the SOL [6]. <br>
    
However due to the short ELM duration of $\sim {200}{\mu s}$ [7] the high heat flux can result in damage of PFCs and diagnostics in future large tokamaks such as ITER [6]. The energy of each ELM and the associated heat load can be reduced by increasing the ELM frequency (ELM mitigation) or by finding mechanisms to prevent ELM instabilities (ELM suppression) [8]. <br>
Both ELM mitigation and ELM suppression require a better understanding of the characteristics of ELMs which is why it is important to have robust detection methods and knowledge of their typical shapes. <br>

A key challenge in detecting ELMs is their different sizes and shapes as well as their quasi-periodicity. The latter does not allow for detection being simply based on a certain ELM frequency $f_{ELM}$. <br>

Gas puff imaging (GPI), being one of the diagnostics used in this report to detect and analyse ELMs, uses neutral gas injection of deuterium or helium to observe the plasma density in the SOL [9] of the tokamak plasma. As the neutral background density is increased by the gas puff, the collision with plasma particles leads to spontaneous emission which is then detected by a fast camera with a $64 \times {64}{px}$ spatial resolution and a ${5}{\mu s}$ frame rate. <br>
</p>

<p>A characteristic of a magnetically confined fusion configuration is its safety factor which is defined as the inverse of the rotational transform $i$
$$
    q \equiv \frac{2 \pi}{i}  \tag{1}
$$
where the rotational transform is given as the change in the poloidal angle $\theta$ after a full toroidal transit of the helical formed magnetic field lanes:
$$
    i(r) \equiv \Delta \theta(r) = \theta(\varphi + 2 \pi R) - \theta(\varphi)    \tag{2}
$$
with the toroidal angle $\varphi$ and the tokamak major radius $R$ [2]. For a torus this becomes
$$
    i(r) = 2\pi \frac{R B_{p}(r)}{r B_{t}(r)}  \tag{3}
$$
with the poloidal magnetic field $B_{p}$ and the toroidal magnetic field $B_{t}$. Hence the safety factor is given by
$$\label{q-value}
    q(r) = \frac{2 \pi}{\Delta \theta(r)} = \frac{r B_{t}(r)}{R B_{p}(r)}    \tag{4}
$$
The first step of equation \ref{q-value} best illustrates that the q-value can be understood as a measure of how many toroidal transits of the magnetic field lanes are necessary in order to achieve a full poloidal transit, i.e $\vartheta = 2 \pi$. 
The name safety factor originates from the fact that plasmas with larger q-values are more stable against magnetohydrodynamic (MHD) instabilities [2]. An often used q-value is $q_{95}$ which represents the q value at $95\%$ of the minor radius $r$ [11,12].

<h4>2.2 ELM frequency</h4>
<p>A mechanism was developed to assign one frequency $f_\mathrm{ELM}$ to a given time history of the GPI or Ipolsola data.
This was being done by calculating the autocorrelation function (ACF) for a given GPI or Ipolsola signal and then taking the frequency that corresponds to the peak in the fourier transform of the ACF as the ELM frequency. <br>
The discrete autocorrelation function of the discrete real function s[t] is given by
$$
    \mathrm{ACF}(s[t])(\tau) =  \sum_{t=0}^{n-1} s[t+\tau] * s[t]   \tag{5}
$$
where $\tau$ was chosen to run between $[-\frac{n-1}{2},\frac{n-1}{2}]$ for odd n and between $[-\frac{n}{2},\frac{n}{2}-1]$ for even n such that the resulting function has a length of n. Note that any terms where one of the indexes runs outside the range of s[t] is set to zero. <br>
The discrete fourier transform (DFT) of the ACF is given by:
$$
    s[k] = \sum_{\tau=0}^{n-1} \mathrm{ACF}[\tau] \exp \left\{-i 2\pi\frac{k}{n}\tau\right\}     \tag{6}
$$
where k=0,..,n-1.</p>

```python
def detectPuffs(arraylength=500,threshold=0.05,FlagDelta=int(0.5/5e-6)):
    t0idx, t1idx = 0, time.shape[0]
    x = arraylength
    f_tilde = []
    for i in range(t0idx,t1idx-2*x):
        y = np.sum(frames[i+x:i+2*x,50,30],axis=0) - np.sum(frames[i:i+x,50,30],axis=0)
        if y>0:
            f_tilde.append(y)
        else:
            f_tilde.append(0)
    maxvalue = np.max(f_tilde)
    f_tilde = [i/maxvalue for i in f_tilde] #normalize f_tilde
    thres = threshold
    f2 = []
    for i in f_tilde:
        if i >= thres:
            f2.append(i)
        else:
            f2.append(0)

    #mark start and final idx of each peak in gradDetectionResult
    truncStart = []
    truncFinal = []
    for i in range(1,len(f2)):
        if (f2[i] != 0 and f2[i-1] == 0):
            truncStart.append(i)
        if (f2[i] == 0 and f2[i-1] != 0):
            truncFinal.append(i)

    if len(truncStart) != len(truncFinal):
        print("error: truncation arrays not same size, check if time series starts or/and ends on peak")
        print("the lengths are: len(truncStart) = "+str(len(truncStart))+"; len(truncFinal) = "+str(len(truncFinal)))
    else:
        #correcting for too many flags in one area which are triggered by "sharp" threshold
        truncStart2 = []
        truncFinal2 = []
        Startidx = []

        for i in range(1,len(truncStart)):
            if (i == 1):
                truncStart2.append(truncStart[0]) #the first start item is valid by definition of this filtering
            if (truncStart[i]-FlagDelta > truncStart[i-1]):
                truncStart2.append(truncStart[i]) #item only appended if it is a "true" start item
                Startidx.append(i) #collection of the used idx of truncStart
        for k in Startidx:
            #the final item before each start item is the final item to the predecessor start item
            truncFinal2.append(truncFinal[k-1])

        truncFinal2.append(truncFinal[-1]) #the last Final item is valid by definition of this filtering
        # *** correction end ***

        if len(truncStart2) != len(truncFinal2):
            print("error: corrected truncation arrays are not the same size")
            print("the lengths are: len(truncStart2) = "+str(len(truncStart2))+"; len(truncFinal2) = "+str(len(truncFinal2)))
        else:
            timeStartResult = [time[i] for i in truncStart2]
            timeFinalResult = [time[i] for i in truncFinal2]
            print(len(truncStart2)," puffs detected with the following start and final times")
            print("Start timestamps: ",timeStartResult)
            print("Final timestamps: ",timeFinalResult)
            return np.array([truncStart2,truncFinal2]) #start and final indices
```

```python
def gradDetection(puff_no,diagnostic,R=55,arraylength=100,deltaIdx=[0,0],FlaggingThreshold=0.25, FlagDelta=300):
    
    validDiagn = {"GPI","Ipolsola"}
    if diagnostic not in validDiagn:
        raise ValueError("results: diagnostic must be one of %r." % validDiagn)
        
    validR = {55,22}
    if R not in validR:
        raise ValueError("results: R must be one of %r." % validR)
    
    if diagnostic == "GPI":
        t0idx, t1idx = puffs[0,puff_no-1]+deltaIdx[0], puffs[1,puff_no-1]+deltaIdx[1]
        if R == 55:
            absGrad = np.absolute(np.gradient(framesAVG_R55[0:t1idx]))
        if R == 22:
            absGrad = np.absolute(np.gradient(framesAVG_R22[0:t1idx]))
    elif diagnostic == "Ipolsola":
        t0idx, t1idx = IpolsolaPuffs[0,puff_no-1]+deltaIdx[0], IpolsolaPuffs[1,puff_no-1]+deltaIdx[1]
        absGrad = np.absolute(Ipolsola["data"][0:t1idx])  #the raw Ipolsola data is already similar to the grad of GPI
    
    # step 1 start
    f_tilde = []
    x = arraylength
    for i in range(t0idx,t1idx-2*x):
        f = np.sum(absGrad[i+x:i+2*x]) - np.sum(absGrad[i:i+x])
        if f>0:
            f_tilde.append(f)
        else:
            f_tilde.append(0)
    maxvalue = np.max(f_tilde)
    f_tilde = [i/maxvalue for i in f_tilde] #normalize f_tilde
    # step 1 end
    
    flags = flagELM(f_tilde=f_tilde,FlaggingThreshold=FlaggingThreshold,FlagDelta=FlagDelta)   
    
    return [flags[0],flags[1],np.add(t0idx+arraylength,flags[2])]

def flagELM(f_tilde,FlaggingThreshold=0.5,FlagDelta=300):
    t0idx = 0
    thres = FlaggingThreshold
    f2 = []
    x = 0 #offset: the 0th conv element is the 0th signal element overlapping woth 0th ELM element
    
    for i in f_tilde:
        if i >= thres:
            f2.append(i)
        else:
            f2.append(0)
    #mark start and final idx of each peak in gradDetectionResult
    truncStart = []
    truncFinal = []
    for i in range(1,len(f2)):
        if (f2[i] != 0 and f2[i-1] == 0):
            truncStart.append(i)
        if (f2[i] == 0 and f2[i-1] != 0):
            truncFinal.append(i)

    if len(truncStart) != len(truncFinal):
        print("error: truncation arrays not same size, check if time series starts or/and ends on peak")
        print("the lengths are: len(truncStart) = "+str(len(truncStart))+"; len(truncFinal) = "+str(len(truncFinal)))
    else:
        #correcting for too many flags in one area which are triggered by "sharp" threshold
        truncStart2 = []
        truncFinal2 = []
        Startidx = []

        for i in range(0,len(truncStart)): #CHANGED
            if (i == 0): #CHANGED
                truncStart2.append(truncStart[i]) #the first start item is valid by definition of this filtering
            else: #CHANGED
                if (truncStart[i]-FlagDelta > truncStart[i-1]):
                    truncStart2.append(truncStart[i]) #item only appended if it is a "true" start item
                    Startidx.append(i) #collection of the used idx of truncStart
        for k in Startidx:
            #the final item before each start item is the final item to the predecessor start item
            truncFinal2.append(truncFinal[k-1])

        truncFinal2.append(truncFinal[-1]) #the last Final item is valid by definition of this filtering
        # *** correction end ***

        if len(truncStart2) != len(truncFinal2):
            print("error: corrected truncation arrays are not the same size")
            print("the lengths are: len(truncStart2) = "+str(len(truncStart2))+"; len(truncFinal2) = "+str(len(truncFinal2)))
        else:
            flagCoords = []
            for j in range(len(truncStart2)):
                flagCoords.append(np.argmax(np.array(f2[truncStart2[j]:truncFinal2[j]]))+x+truncStart2[j]+t0idx)
                #+x to get to middle pixel idx,+truncStart[j] to correct for truncation shift
                #+t0idx to correct for gaspuffshift/overall interval shift
            return [f_tilde,f2,np.array(flagCoords)]
```

In the first step a moving average (MA) with a boxcar length of $\Delta t = {500}{\mu s}$ is created 
$$
    \mathrm{MA}[t] = \sum_{n=0}^{\Delta t} s[t-n]  \tag{7}
$$
with $s[t]$ being the GPI data.
This is then used to compute a new function $f[t]$ for each data point:
$$
    f[t] = \mathrm{MA}[t+\Delta t] - \mathrm{MA}[t] \tag{8}
$$
The idea here is that a local maximum of $f[t]$ means that this point corresponds to the start of an ELM. It is therefore helpful to set all points where the opposite is the case to zero by defining an auxiliary function $\tilde{f}$ 
$$\label{eq:f-tilde}
    \tilde{f}[t] = 
    \begin{cases}
  f[t] & \text{for }f[t] > 0 \\    
  0    & \text{for }f[t]\le 0    \tag{9}
\end{cases}
$$
The normalised output of this step is shown in figure 6.

```python
plt.figure(figsize=(10,6))
plt.plot(np.arange(0,len(f_tilde)*5e-6,5e-6),f_tilde)

plt.xlabel('time [s]',fontsize=14)
plt.ylabel(r'$\tilde{f}\: [t]$',fontsize=14)
plt.title(r'$\tilde{f}\: [t]$ based on 1st time derivative of GPI of shot #36341',fontsize=14)
plt.xlim(0,0.075)
```

<b>Figure 6:</b> The normalised auxiliary function $\tilde{f}$ defined in equation 9 plotted for a segment of the GPI data of #36341 and normalised by the maximum value of $f[t]$.

Now a variable parameter of the algorithm is introduced that will be referred to as the <i>flagging threshold</i>. This parameter excludes some of the visible peaks seen in figure 6 by requiring a minimum height for a peak:
$$\label{eq:f2}
    f_2 = 
\begin{cases}
  \tilde{f}[t] & \text{for }\tilde{f}[t] > \mathrm{threshold} \\  
  0    & \text{for }\tilde{f}[t]\le \mathrm{threshold}              \tag{10}
\end{cases}
$$
In the next step, the algorithm searches for local maxima of $f_2$ by determining the maxima within each time interval during which all data points are consistently different from zero (referred to as <i>search intervals</i>):
$$
    \max(f_2[t]) = f_2[t_{ELM}]   \tag{11}
$$
where t runs only within the i-th search interval $t \in [t_i^{\mathrm{start}},t_i^{\mathrm{final}}]$. <br>
However, by applying the <i>flagging threshold</i> to $\tilde{f}$ there is sometimes more than one search interval within one larger peak that seems to be caused by one ELM only. This phenomenon is displayed in figure 7 and stems from the sharp character of the threshold. Little fluctuations of $\tilde{f}$ around the threshold will cause multiple search intervals.

To correct for this, a second parameter is introduced that will be referred to as the <i>flagging delta</i> and which allows to merge some search intervals together into larger <i>valid intervals</i>. <br>
A search intervals' start time is only a start time for a valid interval if the start time of the last valid interval is further away than <i>flagging delta</i>. 
Identifying a new valid intervals' start time as the start time of the i-th search interval allows to determine the final time of the last valid interval as the final time of the last search interval $i-1$. <br>
In a final step the maxima of each valid interval is determined and the associated time saved as the start of one ELM.

<h4>2.4  Mechanism 2: cross-correlation</h4>
<p> This mechanisms makes use of the cross-correlation of the GPI signal and the time series of a single ELM. Let s[t] be the GPI signal and g[t] the shorter time series of a single ELM region of length $\Delta t = {4.5}{ms}$.
The ELM region sample $g[t]$ has a labelled ELM start time $t_{sample}^{ELM}$. The discrete cross-correlation is then defined as:
$$\label{eq:conv}
    c[\tau] = g[t] * s[t] = \sum_{t=0}^{\Delta t} g[\tau+t] \cdot s[t]      \tag{12}
$$
The idea is that $c[\tau]$ will have higher values for shift times $\tau$ where $s$ has a similar shape to $g$. The output of this step can be seen in figure 9.</p>

The local maxima will then respond to the perfect overlap between both signals. The location of the ELM in $s[t]$ can be calculated as:
$$
    t_i^{elm} = \tau_i^{ELM} + t_{sample}^{ELM}     \tag{13}
$$
where $\tau = 0$ corresponds to the first complete overlap of both $s$ and $g$. As output $c$ in figure 9 has a similar shape as the function $\tilde{f}$ of Mechanism 1 in section 2.3, the remainder of this algorithm works analogous to Mechanism 1 with the use of a <i>flagging threshold</i> and a <i>flagging delta</i>.
Note however, that generally the optimal values for these parameters will be different for both mechanisms.

```python
def convDetection(puff_no,diagnostic,FlaggingThreshold=0.5,FlagDelta=300,deltaIdx=[0,0]):
    
    validDiagn = {"GPI","Ipolsola"}
    if diagnostic not in validDiagn:
        raise ValueError("results: diagnostic must be one of %r." % validDiagn)
        
    if diagnostic == "GPI":
        elm = np.load("data/GPIconvSample.npy")
        flag_idx = 269
        t0idx, t1idx = puffs[0,puff_no-1]+deltaIdx[0], puffs[1,puff_no-1]+deltaIdx[1]
        conv = np.correlate(framesAVG_R55[t0idx:t1idx],elm,mode="full")
    elif diagnostic == "Ipolsola":
        elm = np.load("data/IpolConvSample.npy")
        flag_idx = 200
        t0idx, t1idx = IpolsolaPuffs[0,puff_no-1]+deltaIdx[0], IpolsolaPuffs[1,puff_no-1]+deltaIdx[1]
        conv = np.correlate(Ipolsola["data"][t0idx:t1idx],elm,mode="full") 
        
    conv = np.divide(conv,np.max(conv)) #normalize
    conv_time = np.arange(0,len(conv)*5e-6,5e-6) #create shift time tau array

    flaggedELMs = flagELM(f_tilde=conv,FlaggingThreshold=FlaggingThreshold,FlagDelta=FlagDelta)
    # correcting tau = 0 at puff beginning
    flaggedELMs = np.add(flaggedELMs[2],t0idx)
    
    #offset caused by right edges (re) of elm region lining up giving max in this algorithm
    #therefore correct the shift from right edge to actual flag
    re_idx = len(elm)-1
    ELMregion = [[i-re_idx,i+1] for i in flaggedELMs]
    flaggedELMs = np.add(flaggedELMs,flag_idx-re_idx)
   
    return [flaggedELMs.astype(int),ELMregion]
```

<h4>2.5 Parameter optimisation</h4>
<p>The goal of this step is to find the values of the <i>flagging threshold</i> and <i>flagging delta</i> for each mechanism which result in the highest detection efficiency. In order to find the optimal values, ELM labels were created for two gas puffs of the GPI diagnostic. This corresponds to 56 labelled ELMs during $\sim {0.4}{s}$.</p> 

```python
GPI_labels_puff2, GPI_labels_puff3 = np.load("data/GPI_labels_36341.npy",allow_pickle=True)
Ipolsola_labels_puff2, Ipolsola_labels_puff3  = np.load("data/Ipolsola_labels_36341.npy",allow_pickle=True)
```

<p>In the next step a loss function $L$ is defined which assigns a real number to each pair of <i>flagging threshold</i> and <i>flagging delta</i>:
$$\label{eq:loss}
    L(\mathrm{thres},\mathrm{fdelta}) = fn(\mathrm{thres},\mathrm{fdelta}) +fp(\mathrm{thres},\mathrm{fdelta})  \tag{14}
$$
with the false negative rate defined as:
\begin{equation}\label{eq: fn}
    fn = \frac{FN}{FN+TP}   \tag{15}
\end{equation}
and a modified definition of the false positive rate:
$$ \label{eq: fp adapted}
    fp = \frac{FP}{FP+TP}  \tag{16}
$$
with FP,FN and TP being the number of false positive, false negative and true positive detected ELMs of the 56 labelled ELMs, respectively.
Note that in the usual definition of the false positive rate the denominator would be set to $FP+TN$. However, here $TN \gg TP$ and the second term in equation \ref{eq:loss} would be largely suppressed with the usual definition.  <br>
In a next step, the loss function $L$ was then used to iterate through all possible combinations of $(\mathrm{thres},\mathrm{fdelta})$ for a range of both parameters and comparing the detected ELMs for each specific parameter pair to the existing labels. <br>
For the GPI and Ipolsola the range of the flagging threshold was set to $[0.1,0.7]$ with iteration steps of $0.01$, while the range for the flagging delta was set to $[{0}{ms},{1.5}{ms}]$ with iteration steps of ${0.05}{ms}$.
Parameter values outside of these boundaries showed to have significantly higher loss values $L$. <br>    
A labelled ELM was counted as being correctly detected if the flag raised by the mechanism lies within an interval of $\pm {1}{ms}$ of the label as it showed that flags withing this region seem to refer to the same ELM. It will later be shown that the median period of the ELMs is given by $T_{ELM} \sim {10}{ms}$. Hence this interval corresponds to $\sim 10\%$ of $T_{ELM}$.</p>

```python
def Errorrates(result,groundtruth,AcceptancedeltaIdx=200,withPrint=True):
    gt = groundtruth 
    delta = AcceptancedeltaIdx
    isELM = []
    isNotELM = []
    detected = []
    for i in result:
        isELMtest = False
        k = 0 
        for j in gt: 
            if (abs(i-j)<delta):
                isELMtest = True
                detected.append(j)
                # fix issue of negative fn values by not letting gt item be detected more than once
                gt = np.delete(gt,k)
                break
            k += 1
        if (isELMtest == True):
            isELM.append(i)
        else:
            isNotELM.append(i)
            
    tp = len(isELM) #is ELM and flagged as such
    fp = len(isNotELM) #is not ELM but flagged as such
    fn = len(groundtruth)-len(detected) #is ELM but not flagged as such
    
    if withPrint == True:
        print("TP: ",tp,",FP: ",fp,",FN: ",fn)
    
    fp_ratio = fp/(fp+tp) #false positives / total number of positives (against usual definition)
    fn_ratio = fn/(fn+tp) #false negatives / total number of true ELMs
    
    return [fp_ratio,fn_ratio]

def loss(puff2flags,puff2labels,puff3flags,puff3labels):
    e2 = Errorrates(puff2flags,puff2labels,withPrint=False)
    e3 = Errorrates(puff3flags,puff3labels,withPrint=False)
    fp_avg = np.mean([e2[0],e3[0]])
    fn_avg = np.mean([e2[1],e3[1]])
    print(fp_avg,fn_avg)
    l = fp_avg+fn_avg
    return l

def GPIconvParaOptimizer():
    minLoss = 1
    l = 0
    tresData = []
    for tres in np.arange(0.1,0.7,0.01):
        l += 100/60
        for flagd in np.arange(0,300,10):
            convPuff2flags, convPuff3region = convDetection(puff_no=2,diagnostic="GPI",FlaggingThreshold=tres, FlagDelta=flagd)
            convPuff3flags, convPuff3region = convDetection(puff_no=3,diagnostic="GPI",FlaggingThreshold=tres, FlagDelta=flagd)
            current_loss = loss(convPuff2flags,GPI_labels_puff2,convPuff3flags,GPI_labels_puff3)
            tresData.append(current_loss)
            if current_loss < minLoss:
                minLoss = current_loss
                tresOpt = tres
                flagdOpt = flagd
        print(l,"% done")          
    return [minLoss,tresOpt,flagdOpt,tresData]

def GPIgradParaOptimizer():
    minLoss = 1
    l = 0
    tresData = []
    for tres in np.arange(0.1,0.7,0.01):
        l += 100/60
        for flagd in np.arange(0,300,10):
            gradPuff2flags = gradDetection(puff_no=2,diagnostic="GPI",R=55,FlaggingThreshold=tres, FlagDelta=flagd)
            gradPuff3flags = gradDetection(puff_no=3,diagnostic="GPI",R=55,FlaggingThreshold=tres, FlagDelta=flagd)
            current_loss = loss(gradPuff2flags,GPI_labels_puff2,gradPuff3flags,GPI_labels_puff3)
            tresData.append(current_loss)
            if current_loss < minLoss:
                minLoss = current_loss
                tresOpt = tres
                flagdOpt = flagd
        print(l,"% done")          
    return [minLoss,tresOpt,flagdOpt,tresData]

def IconvParaOptimizer():
    minLoss = 1
    l = 0
    tresData = []
    for tres in np.arange(0.1,0.7,0.01):
        l += 100/60
        for flagd in np.arange(0,300,10):
            convPuff2flags, convPuff2region = convDetection(puff_no=2,diagnostic="Ipolsola",FlaggingThreshold=tres,FlagDelta=flagd,deltaIdx=[IpolsolaDeltaIdxStart,0])
            convPuff3flags, convPuff3region = convDetection(puff_no=3,diagnostic="Ipolsola",FlaggingThreshold=tres,FlagDelta=flagd,deltaIdx=[IpolsolaDeltaIdxStart+500,0])
            current_loss = loss(convPuff2flags,Ipolsola_labels_puff2,convPuff3flags,Ipolsola_labels_puff3)
            tresData.append(current_loss)
            if current_loss < minLoss:
                minLoss = current_loss
                tresOpt = tres
                flagdOpt = flagd
        print(l,"% done")          
    return [minLoss,tresOpt,flagdOpt,tresData]

def IgradParaOptimizer():
    minLoss = 1
    l = 0
    tresData = []
    for tres in np.arange(0.1,0.7,0.01):
        l += 100/60
        for flagd in np.arange(0,300,10):
            gradPuff2flags = gradDetection(puff_no=2,diagnostic="Ipolsola",FlaggingThreshold=tres, FlagDelta=flagd,deltaIdx=[IpolsolaDeltaIdxStart,0])
            gradPuff3flags = gradDetection(puff_no=3,diagnostic="Ipolsola",FlaggingThreshold=tres, FlagDelta=flagd,deltaIdx=[IpolsolaDeltaIdxStart+500,0])
            current_loss = loss(gradPuff2flags,Ipolsola_labels_puff2,gradPuff3flags,Ipolsola_labels_puff3)
            tresData.append(current_loss)
            if current_loss < minLoss:
                minLoss = current_loss
                tresOpt = tres
                flagdOpt = flagd
        print(l,"% done")          
    return [minLoss,tresOpt,flagdOpt,tresData]
```

<h4>2.6 Combining mechanisms</h4>
<p>In a final step, mechanism 1 of section 2.3 and mechanism 2 of section 2.4 were combined for both the GPI and the Ipolsola diagnostic. Not only does this allow to produce a single output of detected ELMs, but also it allows to assign a degree of confidence to each detected ELM. This combined mechanism takes in the detected ELMs of mechanism 1 and mechanism 2 and outputs a list of detected ELM together each with an attribute indicating whether this ELM was detected by both mechanisms, only by mechanism 1 or only by mechanism 2. As in section 2.5 two flags of both algorithms were accepted as referring to the same ELM if they are less than ${1}{ms}$ apart from each other.</p>

```python
def detectCombi(puff_no,diagnostic,R=55,arraylength_grad=100,deltaIdx_grad=[0,0],deltaIdx_conv=[0,0],FlaggingThreshold_grad="optimal", FlagDelta_grad="optimal",FlaggingThreshold_conv="optimal",FlagDelta_conv="optimal",deltaR_conv=17):
    
    validDiagn = {"GPI","Ipolsola"}
    if diagnostic not in validDiagn:
        raise ValueError("results: diagnostic must be one of %r." % validDiagn)

    if diagnostic == "GPI":
        gradOptParaTres, gradOptParaFlagd, convOptParaTres, convOptParaFlagd = 0.27, 260, 0.51, 240

        if FlaggingThreshold_grad == "optimal":
            FlaggingThreshold_grad = gradOptParaTres
        if FlagDelta_grad == "optimal":
            FlagDelta_grad = gradOptParaFlagd
        if FlaggingThreshold_conv == "optimal":
            FlaggingThreshold_conv = convOptParaTres
        if FlagDelta_conv == "optimal":
            FlagDelta_conv = convOptParaFlagd

        gradResult = gradDetection(puff_no=puff_no,diagnostic="GPI",arraylength=arraylength_grad,deltaIdx=deltaIdx_grad,FlaggingThreshold=FlaggingThreshold_grad,FlagDelta=FlagDelta_grad)[2] 
        convResultFlags, convResultRegion = convDetection(puff_no=puff_no,diagnostic="GPI",FlaggingThreshold=FlaggingThreshold_conv,FlagDelta=FlagDelta_conv,deltaIdx=deltaIdx_conv)

    elif diagnostic == "Ipolsola":
        gradOptParaTres,gradOptParaFlagd, convOptParaTres, convOptParaFlagd = 0.18, 290, 0.13, 90 

        if FlaggingThreshold_grad == "optimal":
            FlaggingThreshold_grad = gradOptParaTres
        if FlagDelta_grad == "optimal":
            FlagDelta_grad = gradOptParaFlagd
        if FlaggingThreshold_conv == "optimal":
            FlaggingThreshold_conv = convOptParaTres
        if FlagDelta_conv == "optimal":
            FlagDelta_conv = convOptParaFlagd

        gradResult = gradDetection(puff_no=puff_no,diagnostic="Ipolsola",arraylength=arraylength_grad,deltaIdx=deltaIdx_grad,FlaggingThreshold=FlaggingThreshold_grad,FlagDelta=FlagDelta_grad)[2] 
        convResultFlags, convResultRegion = convDetection(puff_no=puff_no,diagnostic="Ipolsola",FlaggingThreshold=FlaggingThreshold_conv,FlagDelta=FlagDelta_conv,deltaIdx=deltaIdx_conv)

    combiResult = []
    for i in gradResult:
        both = False
        l = 0
        for j,k in zip(convResultFlags,convResultRegion):
            if abs(i-j) < 200:
                both = True
                combiResult.append([i,j,k,"both"])
                convResultFlags = np.delete(convResultFlags,l)
                convResultRegion = np.delete(convResultRegion,l,axis=0)
                break
            l += 1
        if both == False:
            combiResult.append([i,0,[0,0],"grad"])
            
    #append rest of convResultFlags that are not both
    for j,k in zip(convResultFlags,convResultRegion):
        combiResult.append([0,j,k,"conv"])
    
    combiResult = np.array(combiResult,dtype=object)    
    return combiResult
```

Another function was created in order to filter the flags by the detection mechanism behind it. Note that filtering for M1 and M2 also includes the flags that were accepted by CM.

```python
def FilterByFlag(FLagidxes,method="both"):
    if method != "both":
        if method == "grad":
            idx = 0
        if method == "conv":
            idx = 1

        FLagidxes_filtered = []
        for i in FLagidxes:
            if i[idx] != 0 :
                FLagidxes_filtered.append(i)
        FLagidxes_filtered = np.array(FLagidxes_filtered)
    else:
        idx = 0 #added
        FLagidxes_filtered = []
        for i in FLagidxes:
            if i[0] != 0 and i[1] != 0:
                FLagidxes_filtered.append(i)
        FLagidxes_filtered = np.array(FLagidxes_filtered)

    return FLagidxes_filtered[:,idx] #added
```

<h5>3.2.2 Ipolsola and GPI offset</h5>

An offset became apparent between the ELMs detected in the GPI and Ipolsola, showing that the ELM generally occurred later in the Ipolsola than in the GPI. To examine the magnitude of the offset, the time that passed between an Ipolsola flag and the previous GPI flag was calculated. The previous GPI flag was only accepted as such if it happened after the previous Ipolsola flag. 

```python
#first determine natural offset by caused by different start times of diagnostics
GPI_Ipolsola_timeOffset = int(np.rint((time[0]-Ipolsola["time"][0])/5e-6))

time[0], Ipolsola["time"][0], GPI_Ipolsola_timeOffset
```

```python
def IpolsolaOffset(GPIresult,Iresult,timeIdxOffset=0,FlagFilter="grad",GPIFilter="yes",IpolFilter="yes"): #Filter ADDED,searchData=50 deleted
    #determine a start offset to know what order of magnitude to search in
    #also assume that offset is positive, i.e GPI flags before Ipolsola flags
    #also based on fact that Ipolsola sometimes flags before GPI start flagging
    if GPIFilter == "yes":
        GPIidxes = FilterByFlag(Test3GPI,method=FlagFilter)
    elif GPIFilter == "no":
        GPIidxes = GPIresult
        
    if IpolFilter == "yes":
        Iidxes = FilterByFlag(Iresult,method=FlagFilter)
    elif IpolFilter == "no":
        Iidxes = Iresult
        
        
    GPIidxes = np.add(GPIidxes,timeIdxOffset)
    offsetIdx = [] #based on Ipolsola flags
    k = 0
    for i in Iidxes:
        smallerGPIidx = []
        for j in GPIidxes:
            if j<i:
                smallerGPIidx.append(j)
        if k == 0:
            if len(smallerGPIidx) > 0:
                offset = abs(max(smallerGPIidx)-i)
                offsetIdx.append(offset)
            else:
                offsetIdx.append(0)
        else:
            if len(smallerGPIidx) != 0:
                if Iidxes[k-1] > max(smallerGPIidx):
                    offsetIdx.append(0)
                else:
                    offset = abs(max(smallerGPIidx)-i)
                    offsetIdx.append(offset)
            else:
                offsetIdx.append(0)
        k += 1
    offsetIdx = np.array(offsetIdx)
    return offsetIdx
```

In order to reduce the effects of different magnetic locations and therefore different locations within the SOL of the diagnostics, the mean magnetic location of the Ipolsola during each gas puff was calculated. In a next step, the magnetic location of the GPI was averaged along the z-pixel-direction for each pixel in the R-pixel-direction. Note that the z-pixel-direction aligns closely with the surfaces of constant magnetic flux as seen in figure 1. The aligned pixel $R_a$ with the closest magnetic flux value to the average Ipolsola magnetic flux value was chosen for the analysis in the GPI.

```python
def Ipol_rho_mean(shot_no,puff_no):
    divrho = np.load("data/divrho_"+str(shot_no)+"_"+str(puff_no)+"s.npz")
    bgr2_mean_rho = []
    bgr2_std_rho = []
    for i in range(0,16): #baugruppe 2 in figure 2 consists of 16 tiles
        mean_rho = np.mean(divrho["rho2"][:,i]) #avg over time of the gas puff for this tile
        std_rho = np.std(divrho["rho2"][:,i])
        bgr2_mean_rho.append(mean_rho)
        bgr2_std_rho.append(std_rho)
    bgr1_mean_rho = []
    bgr1_std_rho = []
    for i in range(0,22): #baugruppe 1 in figure 2 consists of 22 tiles
        mean_rho = np.mean(divrho["rho1"][:,i]) #avg over time of the gas puff for this tile
        std_rho = np.std(divrho["rho1"][:,i])
        bgr1_mean_rho.append(mean_rho)
        bgr1_std_rho.append(std_rho)
    bgr12_mean_rho = np.append(bgr2_mean_rho,bgr1_mean_rho) #avg over all tiles
    bgr12_std_rho = np.append(bgr2_std_rho,bgr1_std_rho)
    bgr12_totalmean_rho = np.mean(bgr12_mean_rho)
    bgr12_totalstd_rho = np.mean(bgr12_std_rho)
    return [bgr12_totalmean_rho,bgr12_totalstd_rho]

def R_alignCalc(bgr12_totalmean_rho):
    R_aligned = 0
    rho_aligned = []
    rho_delta = 0
    for i in range(0,64): #avg along all 64 R-pixels
        rhoAVG = np.mean(rhoH[i,:]) #avg along z for each r
        rhoSTD = np.std(rhoH[i,:])
        if i == 0:
            rho_delta = abs(bgr12_totalmean_rho - rhoAVG)
        else:
            if abs(bgr12_totalmean_rho - rhoAVG) < rho_delta:
                rho_delta = abs(bgr12_totalmean_rho - rhoAVG)
                rho_aligned = [rhoAVG,rhoSTD]
                R_aligned = i
    return [R_aligned,rho_aligned,rho_delta]
```

Figure 14 that there is a consistent offset between the GPI and Ipolsola of $\Delta t = (2.17 \pm 0.15){ms}$.<br>
The $q_{95}$ value for all shots used for figure 14 is given by $q_{95} \approx 4.8$. The ELM will complete approximately a quarter of a full poloidal transit in order to travel from the GPI at the outer mid-plane to the Ipolsola at the  divertor region as seen in figure REF. Using equation 4, the number of toroidal transits that are necessary to achieve $\Delta \theta = \pi/2$ is given by:
\begin{equation}
    N = \frac{q_{95}}{4} \tag{17}
\end{equation}
Given that the toroidal circumference is given by $C = 2 \pi R$, the toroidal distance travelled by the ELM can be approximated as:
\begin{equation}
    \Delta x_t = N \cdot C = \frac{\pi R q_{95}}{2} \approx {11.31}{m}    \tag{18}
\end{equation}
with $R \approx {1.5}{m}$. The poloidal distance travelled can be approximated by the distance in z-direction between the outer mid-plane and the divertor $\Delta z \approx {1}{m}$ as seen in figure 2.
The total distance can then be approximated by:
\begin{equation}
    \Delta x \approx \sqrt{(\Delta x_t)^2 + (\Delta z)^2} \approx {11.35}{m}   \tag{19}

<h5>3.2.3 ELM frequency & Periodicity</h5>
<p>
Figure 15 shows the time $\Delta t$ that passes between each successive pair of ELMs as well as the mean $\overline{T}_{ELM}$ and standard deviation $\sigma$ for each of the seven included gas puffs for the GPI and the Ipolsola diagnostic. 
The coefficient of variation (CV) is given by
\begin{equation}\label{eq:cv}
    c_v = \frac{\sigma}{\overline{T}_{ELM}} \tag{24}
\end{equation}    
The total mean ELM period is given by $\overline{T}_{GPI} = {9.3}{ms}$ in the GPI and $\overline{T}_{Ipol} = {8.5}{ms}$ in the Ipolsola. The total mean standard deviation are given by $\sigma_{GPI} = {3.4}{ms}$ and $\sigma_{Ipol} = {2.5}{ms}$. Using equation \ref{eq:cv}, this leads to a mean coefficient of variation of $36\%$ for the GPI and $29\%$ for the Ipolsola.
A perfectly periodic behaviour would be associated with a coefficient of variation of zero, whereas aperiodic behaviour, i.e a uniform distribution of $\Delta t$ would be associated with a coefficient of variation of $1/\sqrt{3} \approx 58\%$ in the continuous limit (see Appendix 1).
Therefore, the CV values determined for the GPI and Ipolsola confirm the quasi-periodic behaviour of ELMs stated in the works of Connor [16] and Keilhacker <i>et al.</i> [17].<br>

For the same two puffs there is a strong deviation between $T_{ACF}$ and $T_{ELM}$ in both diagnostics. When looking at the data, these puffs show to be more noisy between ELMs than the other puffs which could explain this deviation. However, this confirms that even though the ELMs show quasi-periodic behaviour, the method described in section 2.2 provides a good approximation for the mean ELM frequency in a given gas puff.</p>

<h4>5.1 Appendix 1 - uniform distribution</h4>
<p>In the continuous limit the probability density function of a uniform distribution that has values different from zero between A and B is given by:
\begin{equation}
    f(x) = \frac{1}{B-A}
\end{equation}
with the mean given by
\begin{equation}
    \mu = \int_{-\infty}^{\infty} f(x)x \, dx  = \frac{B+A}{2}
\end{equation}
The variance is after a rather lengthy calculation [18] given by
\begin{equation}
    \sigma^2 = \langle x^2 \rangle - \mu^2 = \frac{(B-A)^2}{12}
\end{equation}
Hence the coefficient of variation is given by
\begin{equation}
    c_v = \frac{\sigma}{\mu} = \frac{(B-A)/\sqrt{12}}{(B+A)/2} = \frac{B-A}{\sqrt{3}(B+A)}
\end{equation}
In the case where $A = 0$ this simplifies to
\begin{equation}
    c_v = \frac{1}{\sqrt{3}}
\end{equation}
</p>
